 Let's encode using Target Encoding

```python
# First Extracting only the CATEGORICAL COLUMNS
categorical_cols= list( X_train.select_dtypes(exclude=['int64','float64']).columns)
```

```python
#  Again we check the count of null values.
X_train[categorical_cols].isnull().sum().sort_values(ascending = False)
```

```python
# Encoding Categorical columns with TARGET ENCODING
from category_encoders import TargetEncoder

target_encoder = TargetEncoder(cols = categorical_cols , 
                               handle_missing='missing')
X_train = target_encoder.fit_transform(X = X_train , 
                                       y =y_train , 
                                       cols = categorical_cols)
```

* Target encoding is the process of replacing a categorical value with the mean of the target variable while at the same time does not add to the dimensionality of the dataset

```python
#Lets check the feature values again in our training data
X_train
```

* AS seen from the output above all the categorical features have been encoded succesfully.

#### Dealing with Highly correlated data

```python
corr = X_train.corr() # Gives Correlation Matrix
columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
selected_columns = X_train.columns[columns] # The selected Columns
X_train = X_train[selected_columns]  # Update Training Data
```

* In the above code we compared the correlation between features and remove one of two features that have a correlation higher than 0.9

```python
# checking the number of features after dropping correlated features
print("Toatl number of features remaining:",len(selected_columns))
```

```python
# Saving the remaining feature columns
col =list(X_train.columns)
```

```python
# updating out test set to have the same feature columns as the training set
X_test = X_test[col]
```

* So our remaining features after dropping the correlated features is 17

* Since the range of values of our data varies widely, in some machine learning algorithms, objective functions will not work properly without **normalization**. 

* For example, many classifiers calculate the distance between two points by the Euclidean distance. If one of the features has a broad range of values, the distance will be governed by this particular feature. 
* Therefore, the range of all features should be *normalized* so that each feature contributes approximately proportionately to the final distance.

* This phenomenon is called **Feature Scaling**.


```python
# Using Standard Scaler for feaure scalling
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled= pd.DataFrame(scaler.fit_transform(X_train),
                       columns = X_train.columns)  
```

* **Standard Scaler** follows the Standard Normal Distribution, i.e., it assumes a normal distribution for data within each feature. The scaling makes the distribution centered around 0, with a standard deviation of 1 and the mean removed.

```python
# Print the scaled features
X_scaled.describe()
```

### Balancing all classes for improving the model
* Often data sets have classes that are not equally represented.  This class imbalance can be a source of frustration since our model tends to be biased towards the class with more samples.

* A good starting point is to employ sampling methods that alter the training data set to provide more class balance
* As various classes are under represented in our dataset we use OVERSAMPLING to balance all the classes

With over sampling, all of the original data is used plus additional duplicates of the data.  For under-represented classes we randomly re-sample data from these classes and use the copies to balance out the classes.

```python
from collections import Counter
counter = Counter(y_train)
# plot the distribution
plt.bar(counter.keys(), counter.values())
plt.show()
print(counter)
```

* It can be easily seen from the above histogram that this is an __imbalanced dataset__  i.e data is not distributed equally among each classes

```python
# Using SMOTE technique for class balancing

from imblearn.over_sampling import SMOTE
```

```python
# Creating a dictionary that maps class labels to the number of desired examples in each class,
# then specifying this via the “sampling_strategy”
strategy = {1:180 , 2:200 , 6:200 , 3:250} 
smote = SMOTE(random_state=42 , 
              k_neighbors=5 , 
              sampling_strategy=strategy)
```

```python
X_scaled , y_train =smote.fit_resample(X_scaled , y_train)
```

```python
# Checking the distribution again
counter = Counter(y_train)
# plot the distribution
plt.bar(counter.keys(), counter.values())
plt.show()
print(counter)
```

###  As training data is much more balanced then before Oversampling

### As we have done all the pre-processing steps, let's move ahead and start building Machine Learning models and see which one is the most suitable i.e has the highest accuracy.

```python
import sklearn.metrics as metrics
from sklearn.metrics import roc_auc_score, roc_curve , log_loss
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
# Importing the library for the cross validation for the training purposes
from sklearn.model_selection import cross_val_score
```

* **N- Fold Cross-validation** involves partitioning a sample of data into complementary subsets, performing the analysis on one subset (called the training set), and validating the analysis on the other subset (called the validation set or testing set)

* This step is used to tune in Hyper Parameters and see which model yeilds best resutls and the we train our on that model

```python
# Creating Models
dtc = DecisionTreeClassifier(criterion="entropy", max_depth=32) #Decision Tree Classifier
mlp = MLPClassifier(hidden_layer_sizes=(64,64,64)) # Multuple Layer Perceptron
rfc = RandomForestClassifier(random_state=42 ) # Random Forest
svm = SVC() # Suport Vector Machine

models = [dtc , mlp , rfc , svm]
```

```python
for i in models:
    all_accuracies = cross_val_score(estimator=i, X=X_scaled, y=y_train, cv=5)
    print(i ,'----------->', all_accuracies.mean())
    print('\n')
```

* It can be clearly seen that MLP and Random Forest are the best performing models on Cross Validatio, both with accuracy close to 70%.

```python
# Hyper parameter tunning

from sklearn.model_selection import GridSearchCV

parameter_space = {
    'max_iter':[2000],
    'early_stopping':[True, False],
    'hidden_layer_sizes': [(32,32,32), (32,32,32, 32)],
    'activation': ['tanh', 'relu'],
    'validation_fraction':[0.3 , 0.2],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant','adaptive'],
}

mlp_clf = GridSearchCV(MLPClassifier(), 
                       parameter_space, 
                       scoring='accuracy', 
                       n_jobs=-1)
mlp_clf.fit(X_scaled,y_train)
```

### MLP MODEL
* We will use  MLP as it was much easier to tweak rather than Random forest which is like a black box and quite prone to over fitting.


```python
# Loading the model 
mlp = MLPClassifier(hidden_layer_sizes=(32,32,32,32), #Setting hiddenLayers = 32 x 4
                    random_state=42 , 
                    early_stopping=True, # to prevent overfitting
                    max_iter=2000, 
                    validation_fraction=0.3 ,
                    verbose=False)
```

```python
# Training the Classifier
mlp.fit(X_scaled,y_train)
```

```python
# Predicting on Trainset
y_pred = mlp.predict(X_scaled)
```

```python
# Evaluating the Model’s Accuracy
print("Accuracy:",metrics.accuracy_score(y_train,y_pred)) 
```

**Thus we are getting training accuracy of 66%**
